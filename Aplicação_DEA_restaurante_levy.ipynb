Normalizando para melhorar interpretação da distribuição dos dados

```python
# Inicializar o gráfico
fig, axs = plt.subplots(nrows=6, ncols=2, figsize=(15, 15))
axs = axs.flatten()  # Flatten para facilitar o acesso

# Lista de colunas a serem normalizadas
colunas = ['Quantidade', 'Faturamento', 'Preço médio', 'Custo Total', 'Custo unit', 'Lucro']

# Percentis para definir limites
percentil_min = 0.1
percentil_max = 99

# Normalizar e plotar cada coluna
for i, coluna in enumerate(colunas):
    # Obter os valores originais
    valores_orig = df[coluna].values.flatten()

    # Calcular a normalização (valor/max.valor)
    max_val = valores_orig.max()
    valores_norm = valores_orig / max_val

    # Obter os percentis para o histograma original
    min_val_orig = np.percentile(valores_orig, percentil_min)
    max_val_orig = np.percentile(valores_orig, percentil_max)

    # Plotar o histograma original
    sns.histplot(valores_orig, bins=50, kde=True, ax=axs[2*i])
    axs[2*i].set_title(f'Distribuição Original - {coluna}')
    axs[2*i].set_xlim(left=min_val_orig, right=max_val_orig)  # Ajusta o limite do eixo X

    # Obter os percentis para o histograma normalizado
    min_val_norm = np.percentile(valores_norm, percentil_min)
    max_val_norm = np.percentile(valores_norm, percentil_max)

    # Plotar o histograma normalizado
    sns.histplot(valores_norm, bins=50, kde=True, ax=axs[2*i + 1])
    axs[2*i + 1].set_title(f'Distribuição Normalizada (Valor/Max. Valor) - {coluna}')
    axs[2*i + 1].set_xlim(left=min_val_norm, right=max_val_norm)  # Ajusta o limite do eixo X

plt.tight_layout()
plt.show()
```

Definir inputs e outputs

```python
# Agrupar os dados pela coluna 'Grupo' e somar os valores de inputs e outputs
df_grouped = df.groupby('Grupo').sum().reset_index()

# Definir inputs e outputs
inputs = df_grouped[['Custo Total', 'Quantidade']]
outputs = df_grouped[['Faturamento', 'Lucro']]

# Pesos dos inputs e outputs
input_weights = np.array([1, 0.8])
output_weights = np.array([0.5, 1])

inputs, outputs, input_weights, output_weights
```

Normalizando os inputs e outputs

```python
# Normalizar os inputs e outputs dividindo pelo maior valor em cada coluna
inputs_normalized = inputs / inputs.max()
outputs_normalized = outputs / outputs.max()

# Aplicar os pesos nos inputs e outputs normalizados
inputs_weighted = inputs_normalized * input_weights
outputs_weighted = outputs_normalized * output_weights

inputs_normalized, outputs_normalized, inputs_weighted, outputs_weighted
```

Definição das DMUs (Decision Making Units) e cálculo dos scores de eficiência

```python
# Lista para armazenar as DMUs e eficiências
n_dmus = df_grouped.shape[0]
efficiency_scores = []

# Calcular a eficiência para cada DMU
for i in range(n_dmus):
    # Somar os inputs ponderados e outputs ponderados da DMU i
    weighted_sum_inputs = np.sum(inputs_weighted.iloc[i, :])
    weighted_sum_outputs = np.sum(outputs_weighted.iloc[i, :])

    # Calcular a eficiência como a razão entre a soma ponderada dos outputs e dos inputs
    efficiency = weighted_sum_outputs / weighted_sum_inputs
    efficiency_scores.append(efficiency)

# Transformar a lista de eficiências em um array para facilitar o manuseio
efficiency_scores = np.array(efficiency_scores)

efficiency_scores
```

Diagnóstico do ranking de eficiência

```python
# Lista para armazenar as DMUs, eficiências, inputs ponderados e outputs ponderados
n_dmus = df_grouped.shape[0]
efficiency_scores = []
weighted_inputs_list = []
weighted_outputs_list = []

# Calcular a eficiência para cada DMU
for i in range(n_dmus):
    # Somar os inputs ponderados e outputs ponderados da DMU i
    weighted_sum_inputs = np.sum(inputs_weighted.iloc[i, :])
    weighted_sum_outputs = np.sum(outputs_weighted.iloc[i, :])

    # Calcular a eficiência como a razão entre a soma ponderada dos outputs e dos inputs
    efficiency = weighted_sum_outputs / weighted_sum_inputs

    # Armazenar os resultados nas listas
    efficiency_scores.append(efficiency)
    weighted_inputs_list.append(weighted_sum_inputs)
    weighted_outputs_list.append(weighted_sum_outputs)

# Transformar as listas em arrays para facilitar o manuseio
efficiency_scores = np.array(efficiency_scores)
weighted_inputs_list = np.array(weighted_inputs_list)
weighted_outputs_list = np.array(weighted_outputs_list)

# Criar DataFrame com os resultados
df_results = pd.DataFrame({
    'Grupo': df_grouped['Grupo'],
    'Eficiência': efficiency_scores,
    'Weighted Inputs': weighted_inputs_list,
    'Weighted Outputs': weighted_outputs_list
})

# Ordenar as DMUs pelo ranking de eficiência
df_results = df_results.sort_values(by='Eficiência', ascending=False)
df_results['Ranking'] = df_results['Eficiência'].rank(ascending=False, method='min').astype(int)

# Exibir a tabela de resultados com ranking
df_results = df_results.reset_index(drop=True)
df_results
```

Plotagem do gráfico (a refinar)

```python
# Ordenando o DataFrame pela coluna 'Eficiência'
df_results_sorted = df_results.sort_values(by='Eficiência')

# Criando a figura e o eixo
plt.figure(figsize=(12, 8))

# Plotando os inputs ponderados e outputs ponderados
scatter = sns.scatterplot(data=df_results_sorted, x='Weighted Inputs', y='Weighted Outputs', hue='Grupo', palette='dark', s=100)

# Adicionando rótulos de dados
for i, label in enumerate(df_results_sorted['Grupo']):
    plt.text(df_results_sorted['Weighted Inputs'].iloc[i], df_results_sorted['Weighted Outputs'].iloc[i],
             f'{label}\nInputs: {df_results_sorted["Weighted Inputs"].iloc[i]:.2f}\nOutputs: {df_results_sorted["Weighted Outputs"].iloc[i]:.2f}\nEficiência: {df_results_sorted["Eficiência"].iloc[i]:.2f}',
             fontsize=9, ha='center', va='bottom')

# Linha de eficiência de 100%
max_value = df_results_sorted[['Weighted Inputs', 'Weighted Outputs']].max().max()
plt.plot([0, max_value], [0, max_value], 'k--', label='Fronteira CRS (100%)')

# Adicionando a linha pontilhada da Fronteira de Eficiência VRS
vrs_x = [0, 0.2, 1.8]
vrs_y = [0, 0.12, 1.5]
plt.plot(vrs_x, vrs_y, 'r--', label='Fronteira de Eficiência VRS')

# Configurações do gráfico
plt.xlabel('Inputs (normalizados e ponderados)')
plt.ylabel('Outputs (normalizados e ponderados)')
plt.title('Gráfico de Eficiências com Inputs e Outputs')
plt.legend(title='DMUs', loc='upper left')
plt.grid(True)

# Exibindo o gráfico
plt.show()
```
